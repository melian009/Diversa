### DIFFUSION on single networks

Given : $x_i$ as a continuous trait variable assigned to each node (individual) $i = 1,2,...,N$,

$\frac{\textit{d}x_{i}}{\textit{d}t}$ = $\sum_{j}^{N} a_{ij} (x_{j} - x_{i})$

and in matrix form 

$\frac{\textit{d}{\bf x}}{\textit{d}t}$ = $-\mathcal{L}{\bf x}$


${\bf x}$ is the vector describing the dynamical state of each node (individual), i.e., ${\bf x}$ = $(x_{1}, x_{2}, x_{N})^{T}$ 

$\mathcal{L}$ is the Laplacian matrix of the network having elements (as matrix form)

$\mathcal{L}$ = $\mathcal{D}$ - $\mathcal{A}$

where $\mathcal{D}$ is the degree matrix (i.e., the degree of each node in the diagonal and the rest of values equal to zero) and $\mathcal{A}$ is the adjancency matrix (i.e., $a_{ij}$ = 1 or 0). This can also be represented as in cell form as

$L_{ij}$ = $k_{i} \delta_{ij} - a_{ij}$

The dynamics of $\frac{\textit{d}{\bf x}}{\textit{d}t}$ given initial condition ${\bf x}(0)$ = ${\bf x_{0}}$ has the solution

${\bf x(t)}$ = $e^{\mathcal{L}t} {\bf x_{0}}$


```python
## Difusion on single graphs
graphics_toolkit();
available_graphics_toolkits();
loaded_graphics_toolkits();
graphics_toolkit("gnuplot")

addpath(genpath(pwd))
pkg load statistics

N = 20; #Number of nodes
A = zeros(N, N); 
Adj = zeros(N * N, N * N); % The adjacency matrix
size(Adj);


% Use 8 neighbors, and fill in the adjacency matrix (as an example)
dx = [- 1, 0, 1, - 1, 1, - 1, 0, 1];
dy = [- 1, - 1, - 1, 0, 0, 1, 1, 1];
for x = 1:N
    for y = 1:N
        index = (x - 1) * N + y;
        for ne = 1:length(dx)
            newx = x + dx(ne);
            newy = y + dy(ne);
            if newx > 0 && newx <= N && newy > 0 && newy <= N
                index2 = (newx - 1) * N + newy;
                Adj(index, index2) = 1;
            end
        end
    end
end

## COMPUTES THE SOLUTION TO THE DIFFERENTIAL EQUATION
Deg = diag(sum(Adj, 2)); % Compute the degree matrix
L = Deg - Adj; % Compute the laplacian matrix in terms of the degree and adjacency matrices
[V, D] = eig(L); % Compute the eigenvalues/vectors of the laplacian matrix
D = diag(D);

## Initial condition (place a few large positive values around and make everything else zero)
C0 = zeros(N, N);
C0(2:5, 2:5) = 5;
C0(10:15, 10:15) = 10;
C0(2:5, 8:13) = 7;
C0 = C0(:);

C0V = V'*C0; % Transform the initial condition into the coordinate system of the eigenvectors
for t = 0:0.1:1
    % Loop through times and decay each initial component
    Phi(t,1) = t;
    Phi(t,2) = C0V .* exp(- D * t); % Exponential decay for each component
end
size(D)
plot(Phi(:,1),Phi(:,2))
```

### DIFFUSION on multiplex network


Given : $x_i^{[\alpha]}$ as a continuous trait variable assigned to each node (individual) $i = 1,2,...,N_{p}$ in layers $\alpha = 1, 2, ..., N_{a}$ and both intralinks and interlinks can sustain diffusion dynamics, then


$\frac{\textit{d}x_{i}^{[\alpha]}}{\textit{d}t}$ = $\mathcal{D}^{[\alpha]}$ $\sum_{j=1}^{N_{p}} w_{ij}^{[\alpha]} (x_{j}^{[\alpha]} - x_{i}^{[\alpha]})$ + $\sum_{k=1}^{N_{a}} \mathcal{D}^{[\alpha,\beta]} (x_{k}^{[\beta]} - x_{i}^{[\alpha]})$

and in matrix form 

$\frac{\textit{d}{\bf X}}{\textit{d}t}$ = $-\mathcal{L}{\bf x}$,

where $\mathcal{L}$ is an ($N_{p}$ $\cdot$ $N_{a}$) x ($N_{p}$ $\cdot$ $M_{a}$) matrix called the $supra$-$Laplacian$ $matrix$, and ${\bf X}$ is an $N_{p}$ $\cdot$ $N_{a}$ vector describing the dynamical state of each node in the multilayer network.
